```python
import pandas as pd

# Crear datos de tiempo y desplazamiento
tiempo = [0, 10, 20, 30, 40, 50]  # en segundos
desplazamiento = [30, 52, 38, 0, -37, -53]  # en metros

# Crear un DataFrame de pandas
datos = pd.DataFrame({'Tiempo (s)': tiempo, 'Desplazamiento (m)': desplazamiento})

# Mostrar el DataFrame
datos
import matplotlib.pyplot as plt

# Datos de tiempo y desplazamiento
tiempo = [0, 10, 20, 30, 40,50]  # en segundos
desplazamiento = [30, 52, 38, 0, -37, -53]  # en metros

# Graficar desplazamiento en función del tiempo con color verde
plt.plot(tiempo, desplazamiento, marker='o', linestyle='-', color= 'green')
plt.title('Gráfica de Desplazamiento en Función del Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Desplazamiento (m)')
plt.grid(True)
plt.show()
```
